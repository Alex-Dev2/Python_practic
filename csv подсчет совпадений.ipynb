```python
import csv
male = 0
female = 0
vse = 0
with open('StudentsPerformance.csv', 'r') as f:
    reader = csv.reader(f, delimiter=';')
    for row in reader:
      for pol in row:
        if (pol.count('female')) == 1:
          female += 1
          vse += 1
        if(pol.count('female')) == 0:
          male += 1
          vse += 1
print(male-1)
print(female)
```
