```python
proverb=input().lower()
new_proverb=''
alfavit='абвгдеёжзийклмнопрстуфхцчшщъыьэюя'
i=0
while i<len(proverb):
    if proverb[i] in alfavit and proverb[i+1] in alfavit:
        new_proverb += proverb[i+1]
        new_proverb += proverb[i]
        i=i+2
    else:
        new_proverb += proverb[i]
        i=i+1
print(new_proverb.capitalize())
```
